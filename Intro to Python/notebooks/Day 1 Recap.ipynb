```python
import numpy as np
```

### Placeholders

```python
np.zeros((3,4))                   #Create an array of zeros
np.ones((2,3,4),dtype=np.int16)   #Create an array of ones
np.empty((3,2))                   #Create an empty array
np.arange(10,25,5)                #Create an array of evenly spaced values (step value)
np.linspace(0,2,9)                #Create an array of evenlyspaced values (number of samples)
np.full((2,2),7)                  #Create a constant array

np.random.random((2,2))           #Create an array with random values
np.random.randint(1,100,6)        #Create an array with random integers

np.random.seed(1234)             # Seed ensures we get the same random numbers
```

### Accessing data in the array

```python
np.random.seed(1234)         # Seed ensures we get the same random numbers
a = np.random.randint(1,100,6)  
b = np.random.randint(1,100,6)
c = np.random.randint(1,100,6)
d = np.random.randint(1,100,6)
e = np.random.randint(1,100,6)

f = np.array([a,b,c,d,e])
f
```

```python
f[1][1]
```

```python
f[1,1]
```

```python
f[2:4,1:4]
```

### Matplotlib - Figures and Axes

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
fig = plt.figure(facecolor='white')
ax1 = fig.add_axes([0,0,1,1])  # left, bottom, width, height
ax2 = fig.add_axes([0.15, 0.1, 0.7, 0.3])
```

```python

fig = plt.figure(facecolor='white')
ax1 = fig.add_axes([0,0,1,1])  # left, bottom, width, height
ax2 = fig.add_axes([0.15, 0.1, 0.7, 0.3])
# Plot the data
ax1.bar([1,2,3],[3,4,5], color = 'blue')
ax2.barh([0.5,1,2.5],[0,1,2], color = 'yellow')
plt.show()
```
